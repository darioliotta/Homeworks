```python
import matplotlib.pyplot as plt
import csv
from scipy import stats
import numpy as np
import sklearn as sl
from sklearn import linear_model
```

```python
# Load the provided data file with the used car data (you can also have a look at it with any text editor)

filename = "data/km_year_power_price.csv"
lines = csv.reader(open(filename, newline=''), delimiter=',')

km = []
year = []
engine = []
price = []

first = 0

for l in lines:
    if (first == 0):
        first = 1
    else:
        km.append(int(l[0]))
        year.append(int(l[1]))
        engine.append(int(l[2]))
        price.append(float(l[3]))
```

Use linear regression to estimate the car prices from the year, kilometers or engine power. 
You can make a simple 1D regression from each one of the parameters independently 




```python
# linear regression with linregress (estimate price from year)

year_values = [float(stats.linregress(year, price).slope), float(stats.linregress(year, price).intercept), float(stats.linregress(year, price).rvalue)]
year_fitted = [year_values[0]*x + year_values[1] for x in year]
```

```python
# Plot the data and the lines representing the output of the linregress and least square algorithms

plt.plot(year, price, 'o', label='original data')
plt.plot(year, year_fitted, 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
# linear regression with linregress (estimate price from power)

engine_values = [float(stats.linregress(engine, price).slope), float(stats.linregress(engine, price).intercept), float(stats.linregress(engine, price).rvalue)]
engine_fitted = [engine_values[0]*x + engine_values[1] for x in engine]

plt.plot(engine, price, 'o', label='original data')
plt.plot(engine, engine_fitted, 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
# linear regression with linregress (estimate price from km)

km_values = [float(stats.linregress(km, price).slope), float(stats.linregress(km, price).intercept), float(stats.linregress(km, price).rvalue)]
km_fitted = [km_values[0]*x + km_values[1] for x in km]

plt.plot(km, price, 'o', label='original data')
plt.plot(km, km_fitted, 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
# Have a look at the correlation coefficients to see which of the 3 features works better

r = {'year' : year_values[2], 'power' : engine_values[2], 'km' : km_values[2]}

for coefficient, value in r.items():
    print("Correletion coefficient for {} is {}".format(coefficient, value))

closest_key, closest_value = min(r.items(), key=lambda item: abs(item[1] - 1))

print("\n")
print(f"The closest coefficient to 1 in modulus is {closest_value} from {closest_key} data")
```
